# 1. Data Cleaning

The raw dataset contained:

- Missing customer IDs
- Missing product descriptions
- Duplicate transactions
- Invalid quantities
- Invalid unit prices

Cleaning steps performed:

- Removed duplicate rows
- Removed transactions with negative quantity
- Removed transactions with zero or negative prices
- Removed rows without customer ID
- Created revenue column
- Converted date columns into usable format

```python
# creating revenue column for business analysis
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
df.head()
```

# 2. Loading Data into PostgreSQL

After cleaning, the dataset was loaded into PostgreSQL.

PostgreSQL was used for:

- Business queries
- Customer analysis
- Sales analysis
- Aggregations
- Creating views for visualization

The cleaned dataframe was stored as the `sales` table.

```python
# connecting to postgreSQL for business analysis
from sqlalchemy import create_engine

username = "postgres"
password = "benedetta"
host = "localhost"
port = "5433"
database = "ecommerce"

engine_url = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_url)

#Loading dataframe into PostgreSql
table_name = "sales"
table_name1 = "original_sales"
df_org.to_sql(table_name1, engine, if_exists="replace", index=False)

from sqlalchemy import text
with engine.connect() as conn:
    conn.execute(
        text("""
    DROP VIEW IF EXISTS
    kpi_summary,
    monthly_revenue,
    monthly_orders,
    top_products_revenue,
    top_products_quantity,
    top_customers,
    rfm_segments,
    customer_retention,
    country_revenue,
    country_orders,
    product_pareto
    CASCADE;
    """)
    )
    conn.commit()
df1.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded to {table_name1} in database: {database}")
print(f"Data successfully loaded to {table_name} in database: {database}")
```
